```python
df = pd.read_sql('''
                SELECT c.movie_id,m.title,m.overview,m.genres,m.keywords,c.cast,c.crew FROM movies as m
                JOIN credits as c on m.title = c.title
                 ''',db)
```

```python
def generos(fila):
    lista =[]
    fila_dic = json.loads(fila)
    for genres in fila_dic:
        lista.append(genres['name'].replace(" ",""))
    return lista
```

```python
generos(df['genres'][100])
```

```python
df['genres'] = df['genres'].apply(generos)
```

```python
df
```

```python
df['keywords'] = df['keywords'].apply(generos)
```

```python
df
```

```python
def elenco(fila):
    lista =[]
    fila_dic = json.loads(fila)
    largo = len(fila_dic)
    if largo > 3:
        largo = 3
    for i in range(largo):
        lista.append(fila_dic[i]['name'].replace(" ",""))
    return lista
```

```python
df['cast'] = df['cast'].apply(elenco)
```

```python
df
```

```python
df['crew'][0]
```

```python
def director(fila):
    fila_dic = json.loads(fila)
    for elemento in fila_dic:
        if elemento['job'] == 'Director':
            return elemento['name']
    return None
    
```

```python
df['crew'] = df['crew'].apply(director)
df
```

```python
df['tags'] = df['overview']+str(df['genres'])+str(df['keywords'])+str(df['cast'])+df['crew']
df.tags
```

```python
df = df.dropna()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer()

matriz_tfidf = vector.fit_transform(df['tags'])
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=6,metric='cosine')
model.fit(matriz_tfidf)
```

```python
def peliculas_similares(titulo_de_pelicula):
    indice = df[df['title'] == titulo_de_pelicula].index[0]
    distances, indices = model.kneighbors(matriz_tfidf[indice])
    peliculas_similares = [(df['title'][i]) for _,i in enumerate(indices[0])]
    return peliculas_similares[1:]
```

```python
peliculas_similares('''Pirates of the Caribbean: At World's End''')
```
